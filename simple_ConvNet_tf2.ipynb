```python
from __future__ import absolute_import, division,print_function, unicode_literals
import numpy as np
import tensorflow as tf
from tensorflow import keras as k
print(tf.__version__)
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
    # Currently, memory growth needs to be the same across GPUs
    for gpu in gpus:
        tf.config.experimental.set_memory_growth(gpu, True)
    logical_gpus = tf.config.experimental.list_logical_devices('GPU')
    print(len(gpus), "Physical GPUs,", len(logical_gpus), "Logical GPUs")
    except RuntimeError as e:
    # Memory growth must be set before GPUs have been initialized
    print(e)
```

```python
fashion_mnist = k.datasets.fashion_mnist
(train_X, train_Y), (test_X, test_Y) = fashion_mnist.load_data()
```

```python
print('The Shape of Training Dataset: {}'.format(train_X.shape))
print('Labels in Training Dataset: {}'.format(len(train_Y)))
print('The Shape of Test Dataset: {}'.format(test_X.shape))
print('Labels in Test Dataset: {}'.format(len(test_Y)))
```

```python
train_X = train_X / 255.0
test_X = test_X / 255.0
```

```python
train_X = tf.expand_dims(train_X, 3)
test_X = tf.expand_dims(test_X, 3)
```

```python
npx = train_X.shape[1]
print('The Shape of Training Dataset: {}'.format(train_X.shape))
print('Labels in Training Dataset: {}'.format(len(train_Y)))
print('The Shape of Test Dataset: {}'.format(test_X.shape))
print('Labels in Test Dataset: {}'.format(len(test_Y)))
```

```python
model = k.models.Sequential()
model.add(k.layers.Conv2D(filters= 50, kernel_size= (3, 3), activation= 'relu', input_shape= (npx, npx, 1)))
model.add(k.layers.MaxPooling2D((2, 2)))
model.add(k.layers.Flatten())
model.add(k.layers.Dense(10, activation= 'softmax'))
model.summary()
```

```python
model.compile(optimizer= 'adam', loss= 'sparse_categorical_crossentropy', metrics= ['accuracy'])
```

```python
model.fit(train_X, train_Y, epochs = 10)
```

```python
train_loss, train_accuracy = model.evaluate(train_X, train_Y)
```

```python
print('train accuracy:', round(train_accuracy * 100, 2), '%')
```

```python
test_loss, test_accuracy = model.evaluate(test_X, test_Y)
```

```python
print('test accuracy:', round(test_accuracy * 100, 2), '%')
```
